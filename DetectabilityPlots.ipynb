```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
hlvk = np.load('Oli/gw_snr_hlvk.npy')
hlv = np.load('Oli/gw_snr_hlv.npy')
hl = np.load('Oli/gw_snr_hl.npy')

l_plus = np.load('Oli/gw_snr_Lplus.npy')
voyage = np.load('Oli/gw_snr_Lvoyager.npy')
ET = np.load('Oli/gw_snr_ET.npy')


afterglow_SKA_tophat_1e_5 = np.load('afterglow_SKA_tophat_n=1e-5.npy')
afterglow_SKA_tophat_1e_3 = np.load('afterglow_SKA_tophat_n=1e-3.npy')
afterglow_SKA_tophat_1e_1 = np.load('afterglow_SKA_tophat_n=1e-1.npy')

afterglow_SKA_gaussian_1e_5 = np.load('afterglow_SKA_gaussian_n=1e-5.npy')
afterglow_SKA_gaussian_1e_3 = np.load('afterglow_SKA_gaussian_n=1e-3.npy')
afterglow_SKA_gaussian_1e_1 = np.load('afterglow_SKA_gaussian_n=1e-1.npy')

afterglow_AMI_gaussian_1e_3 = np.load('afterglow_AMI_gaussian_n=1e-3.npy')
afterglow_AMI_gaussian_1e_5 = np.load('afterglow_AMI_gaussian_n=1e-5.npy')
afterglow_AMI_gaussian_1e_1 = np.load('afterglow_AMI_gaussian_n=1e-1.npy')

afterglow_AMI_tophat_1e_3 = np.load('afterglow_AMI_tophat_n=1e-3.npy')


kilonovae_gw170817_i = np.load('fluxes_i.npy')
kilonovae_gw170817_g = np.load('fluxes_g.npy')
kilonovae_gw170817_r = np.load('fluxes_r.npy')

pulsar_emission = np.load('Pulsar_emission_simple_pulsar_large.npy')
curv_emission = np.load('Curv_emission_simple_pulsar_large.npy')

plt.rcParams["font.family"] = "Times New Roman"
```
